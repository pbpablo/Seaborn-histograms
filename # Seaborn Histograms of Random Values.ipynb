# Seaborn Histograms of Random Values

    

```python
import numpy as np 
import random
#The function randint() generates random integers for you. If you call the function, it returns a random integer N such that a <= N <= b.
#The randint() method to generates a whole number (integer). You can use randint(0,50) to generate a random number between 0 and 50.
numbers = [random.randint(1,25) for i in range (1,1000)] 
```

```python
numbers
```

```python
%matplotlib inline  #%matplotlib inline will make your plot outputs appear and be stored within the notebook.
```

```python
import seaborn as sns  # Why is seaborn imported as sns and not sbn? Answer:Samuel Norman "Sam" Seaborn is a fictional character #portrayed by Rob Lowe on the television serial drama The West Wing.
#So, it's a joked initialism.(Source: Stack overflow)
Values, frequency = np.unique(numbers, return_counts=True)
```

```python
Values #This will display all the random numbers that were generataed from 0-25
```

```python
frequency #This will give a count of the frequncy of occurence of numbers. i.e, number 1 is generataed 36 times.
```

# Uniformly distributed random integer values

```python
sns.set_style('whitegrid') #Sets the background to be Whitegrid
axes = sns.barplot(x=Values, y=frequency, color='green')  #A barplot is basically used to aggregate the categorical data according to some methods and by default it's the mean. To use this plot we choose a categorical column for the x-axis and a numerical column for the y-axis
axes.set_title('Uniform Distribution Using Random Integer Values', color='black') #Sets the title of the plot 
axes.set(xlabel='Numbers', ylabel='Frequency')
```
